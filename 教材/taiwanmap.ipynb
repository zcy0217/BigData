```
#載入所需套件
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
#設定圖型大小
plt.rcParams['figure.figsize']=(10,10)
```

```
# from google.colab import drive
# drive.mount('/content/drive')
```

```
#讀取台灣城鎮資訊及讀取中文字型(utf-8)
town_TA = gpd.read_file('TOWN_MOI_1081121.shp', encoding='utf-8')
```

```
town_TA.head()
```

```
#設定經緯度，繪出台灣地圖
x=town_TA.plot(column='COUNTYNAME',cmap='Blues') #純粹藍色
x.set_xlim(119.7, 122.3) 
x.set_ylim(21.8, 25.4)
plt.title("Taiwan map",fontsize=20)
```

```
#以COUNTYNAME指定為臺南市、繪製臺南地圖，RdBu：紅藍色
tainan_town=town_TA[town_TA['COUNTYNAME']=='臺南市']
tainan_town.plot(cmap="Reds") #純粹紅色
```

```
#載入各縣市染疫人數
data=pd.read_csv('exam1.csv')
```

```
data.head()
```

```
#將各縣市染疫人數合併至台灣地圖資料，以城市名為合併依據(檔一:城市名變數:COUNTYNAME，對應town_TA、檔二:城市名變數:AA，對應data)
town_A_no = town_TA.merge(data, left_on=('COUNTYNAME'), right_on=('AA'))
```

```
town_A_no.head()
```

```
# Colab 進行matplotlib繪圖時顯示繁體中文
# 下載台北思源黑體並命名taipei_sans_tc_beta.ttf，移至指定路徑
!wget -O taipei_sans_tc_beta.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download
!mv taipei_sans_tc_beta.ttf /usr/local/lib/python3.6/dist-packages/matplotlib//mpl-data/fonts/ttf

from matplotlib.font_manager import FontProperties
import matplotlib.pyplot as plt 
```

```
# 自定義下載的字型為myfont
myfont = FontProperties(fname=r'/content/taipei_sans_tc_beta.ttf')

# !!!!後續在相關函式中增加fontproperties=myfont屬性即可!!!!
```

```
#地圖顏色依確診數NO來繪圖
SA=town_A_no.plot(column='NO',cmap='Reds')
SA.set_xlim(119.5, 122.5)
SA.set_ylim(21.5, 25.5)
#設定圖型標題，並以myfont字型呈現標題字型，字體大小為20
plt.title("台灣各縣市疫情分佈",fontproperties=myfont,fontsize=20)
```

```
import pandas as pd
import numpy as np
```

```
#讀取檔案，將檔案名稱設為data
from google.colab import files
df1=pd.read_excel('Mer1.xlsx')
df2=pd.read_excel('Mer2.xlsx')
```

```
df1.head(6)
```

```
df2.head(6)
```

# 一、merge合併模式

```
#inner交集合併模式, on:依'key'欄位合併
df1.merge(df2,on = 'Number',how = 'inner')#內連線，取交集，共同擁有的
```

```
#outer交集合併模式, on:依'key'欄位合併
df1.merge(df2,on = 'Number',how = 'outer')#取聯集，全部都取
```

```
#left合併模式, on:依'Number'欄位合併
df1.merge(df2,on = 'Number',how = 'left')#以左側檔案為基準
```

```
#right合併模式, on:依'Number'欄位合併
df1.merge(df2,on = 'Number',how = 'right')#以右側檔案為基準
```

```
#讀取檔案，將檔案名稱設為data
from google.colab import files
data1=pd.read_excel('2019STA.xlsx')
data2=pd.read_excel('2019STA_2.xlsx')
```

```
data1.head()
```

```
data2.head()
```

```
#不同欄位名稱合併模式, on:依同資料的不同欄位名稱合併
data1.merge(data2,left_on = 'No',right_on='Number')#內連線，取交集,默認取交集
```

# 二、concat合併模式

```
#讀取檔案，將檔案名稱設為data
from google.colab import files
df1=pd.read_excel('Mer1.xlsx')
df3=pd.read_excel('Mer3.xlsx')
```

```
df1.head(6)
```

```
df3.head(6)
```

```
#concat:連結;axis=0直行合併
pd.concat([df1, df3], axis=0)
```

```
#axis=1橫向合併，取inner行取交集
pd.concat([df1, df2], axis=1)
```

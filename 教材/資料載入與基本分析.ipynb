# 單一數量資料分析：平均數、中位數、變異數、標準差

```
#計算「Final_Exam」欄位資料之平均數
data["Final_Exam"].mean()
```

```
#計算「Final_Exam」欄位資料之中位數
data['Final_Exam'].median()
```

```
#計算「Final_Exam」欄位資料之變異數
data["Final_Exam"].var()
```

```
#計算「Final_Exam」欄位資料之標準差
data["Final_Exam"].std()
```

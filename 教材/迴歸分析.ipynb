# 一、線性迴歸模式：解釋分析模式
用於分析哪些自變數(因素)會影響到某個依變數之統計分析模式，其中依變數需為“屬量”的變數，自變數亦為屬量變數

    可參考https://www.statsmodels.org/stable/api.html

```
X=df[["education_years", "working_years"]] #自變數
Y=df["salary"]
```

```
#語法一
import statsmodels.api as smf
XX=smf.add_constant(X) #截句
RA=smf.OLS(Y,XX).fit()
print(RA.summary())
```

F-statistic(F值):101.6

Prob(P值):7.16e-11(趨近於0，小於0.05有顯著相關)  

R-squared:0.914(解釋能力很好，因為趨近於1)

education_years的T值大於working_years，代表更有顯著相關

```
#語法一
#進行迴歸分析，跑出迴歸分析解釋分析表(從statsmodels.formula.api 匯入ols線性迴歸分析)
#PTS為依變數；A, OEFF,TO為自變數，進行迴歸分析
import statsmodels.formula.api as smf
model = smf.ols("salary ~ education_years+ working_years ", df).fit()
print(model.summary())
```

# 二、線性迴歸模式：預測分析模式_機器學習
利用線性迴歸模型，建立依變數之預測模型

```
#2.用線性回歸做訓練線性模型:
#import基本函式庫
import pandas as pd
import numpy as np

#準備X & y array
X = df[["education_years","working_years"]] #education_years, working_years為自變數,可自行調整
y = df['salary']            #salary為依變數,可自行調整

#將資料分成訓練組及測試組
from sklearn.model_selection import train_test_split

#test_size代表測試組比例。random_state代表設定隨機種子，讓測試結果可被重複
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=3)

#載入線性迴歸，並訓練模型
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)

#取得截距。如果公式是y=a+bx，a即是截距
print(lm.intercept_)
print(lm.coef_)

#取得迴歸係數，並用Data Frame顯示:pd.DataFrame(資料,列的名稱,行的名稱)
#lm.coef_
#X_train.columns
cdf = pd.DataFrame(lm.coef_,X_train.columns,columns=['Coef'])
cdf
```

計算模式績效:RMSE, R_Square

```
#計算R_Square,
R2_train = lm.score(X_train,y_train)
R2_test = lm.score(X_test,y_test)

print('train_R2=%.3f, test_R2=%.3f'%(R2_train,R2_test))
```

```
from sklearn.metrics import mean_squared_error
from math import sqrt
# 使用X_test預測y，得到y_pred1 
y_pred1 = lm.predict(X_test)
rmse = sqrt(mean_squared_error(y_test, y_pred1)) 
print("rmse: %.3f" % rmse) 
```

# 練習題：請以nba2.xlsx之資料，以1Q＋2Q＋3Q, A, TOT, TO等因子建立PTS之線性迴歸模式，並說明分析結果。

```
df2=pd.read_excel('nba2.xlsx')
df2['TQ']=df2['1Q']+df2['2Q']+df2['3Q']
```

```
X=df2[["TQ",'A','TOT','TO']] #自變數
Y=df2["PTS"]
```

```
import statsmodels.api as smf
XX=smf.add_constant(X) #截句
RA=smf.OLS(Y,XX).fit()
print(RA.summary())
```

F-statistic(F值):1717

Prob(P值):0(小於0.05有顯著相關)

R-squared:0.737(解釋能力好，因為趨近於1)

TQ:1Q+2Q+3Q的總和

T值的絕對值排序:TQ>A>TOT>TO，代表1Q+2Q+3Q的總和比其他三個更有顯著差異，TO相對較低

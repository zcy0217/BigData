# t檢定:用於比較兩母體的平均數是否有顯著差異(類*量)
T值大P值小 reject H0

T值小P值大 accept H0

```
#t檢定:用於比較兩樣本的平均數是否有顯著差異(類*量)
#匯入scipy模組
from scipy import stats
#語法:stats.ttest_ind(樣本1,樣本2)
#取當W/L=0時的3P與W/L=1時的3P去做t檢定
(t,p)=stats.ttest_ind(df2['PTS'][df2['W/L']==0],df2['PTS'][df2['W/L']==1])
print("t_value=%.3f, p_value=%.3f"%(t,p))
```

```
#列出勝與負球隊在三分球進球數敘述統計的差異
a=df2.groupby('W/L').mean()
b=a[["PTS"]]
b
```

```
#把自變數放進名為NBA_test2的陣列中
NBA_set2=['1Q','2Q', '3Q', '3P', 'FT', 'TOT','A', 'ST', 'TO']
```

```
name=[]
alist=[]
pvlist=[]
for i in range(len(NBA_set2)):
  #分別取NBA_test2變數與W/L=0及W/L=1的交集去做t檢定
  (a,pv)=stats.ttest_ind(df2[NBA_set2[i]][df2['W/L']==0],df2[NBA_set2[i]][df2['W/L']==1])
  #將t_test項目加進name中
  name.append(NBA_set2[i])
  #將a(t值)加進alist中
  alist.append(a)
  #將pv值加進pvlist中
  pvlist.append(pv)
```

```
#lambda表示式是一種效能比較好的迴圈,跑比較快
pd.set_option('display.float_format',lambda x:'%.2f'%x)
#將alist資料結構用DataFrame顯示出
result=pd.DataFrame(alist,name,columns=['t_value'])
#將pvlist加進result第二欄
result.insert(1,'p_value',pvlist)
result
```

# 練習一：請分析東西區(W/E)球隊在得分(PTS)、三分球進球數(3P)、助攻(A)、進攻籃板(OR)、防守籃板(DR)的差異。分析後請說明哪些指標有顯著的差異。

```
a=df2.groupby('W/E').mean()
a[['PTS','3P','OR','DR','A']]
```

```
a=df2.groupby('W/E').std()
a[['PTS','3P','OR','DR','A']]
```

```
NBA_set2=['PTS','3P','OR','DR','A']
name=[]
alist=[]
pvlist=[]
for i in range(len(NBA_set2)):
  (a,pv)=stats.ttest_ind(df2[NBA_set2[i]][df2['W/E']==1],df2[NBA_set2[i]][df2['W/E']==2])
  name.append(NBA_set2[i])
  alist.append(a)
  pvlist.append(pv)
pd.set_option('display.float_format',lambda x:'%.2f'%x)
result=pd.DataFrame(alist,name,columns=['t_value'])
result.insert(1,'p_value',pvlist)
result
```

PTS和3P的P值都小於0.05，所以他們對於W/E有顯著相關，是有顯著差異

其中PTS的T值取絕對值大於3P的T值取絕對值，所以PTS比3P更有顯著差異

OR和DR以及A的P值因為大於0.05所以和W/E沒有顯著差異

# 變異數分析(ANOVA)

```
from scipy import stats
(f,pv)=stats.f_oneway(df2['PTS'][df2['REST DAYS']==0],
                          df2['PTS'][df2['REST DAYS']==1],
                          df2['PTS'][df2['REST DAYS']==2],
                          df2['PTS'][df2['REST DAYS']==3])
print('%.2f,%.2f'%(f,pv))
```

```
a=df2.groupby('REST DAYS').mean()
b=a['PTS']
b
```

```
#無母數統計分析(Kruskal-Wallis Test):當資料無法滿足常態分配或變異數同值性時，不宜採用變異數分析，應採用無母數統計分析
from scipy import stats
(f,pv)=stats.kruskal(df2['PTS'][df2['REST DAYS']==0],
                          df2['PTS'][df2['REST DAYS']==1],
                          df2['PTS'][df2['REST DAYS']==2],
                          df2['PTS'][df2['REST DAYS']==3])
print('%.2f,%.2f'%(f,pv))
```

```
#變異數分析:比較多組樣本之間的平均數差異(類*量)
##把變數放進名為one的陣列中
NBA_set3=['FG', '3P','TOT', 'A', 'ST', 'TO', 'PTS']
name2=[]
flist2=[]
pvlist2=[]
for i in range(len(NBA_set3)):
  #分別取t_test變數與REST DAYS的交集去做變異數分析
  (f,pv2)=stats.f_oneway(df2[NBA_set3[i]][df2['REST DAYS']==0],
                          df2[NBA_set3[i]][df2['REST DAYS']==1],
                          df2[NBA_set3[i]][df2['REST DAYS']==2],
                          df2[NBA_set3[i]][df2['REST DAYS']==3])
  name2.append(NBA_set3[i])
  flist2.append(f)
  pvlist2.append(pv2) 
```

```
result2=pd.DataFrame(flist2,name2,columns=['F_value'])
result2.insert(1,'p_value',pvlist2)
result2
```

# 單一類別資料分析：次數分配與相對次數分配

```
#使用groupby將資料以主客場分組(H/R)
f=data.groupby('H/R')
#用size()看f裡面每個組別內的大小
frequence=f.size()
#將frequence做成dataframe形式，並將資料columns命名為Frequence_主客場
flist=pd.DataFrame(frequence,columns=["Frequence_主客場"])
flist
```

```
#使用groupby將資料以主客場分組(H/R)
f=data.groupby('H/R')
percentage=f.size()/len(data)
#將percentage做成dataframe形式，並將資料columns命名為Percentage_主客場
plist=pd.DataFrame(percentage,columns=["Percentage_主客場"])
plist
```

```
#利用merge功能合併flist與plist兩個陣列
flist.merge(plist,on = 'H/R')
```

```
#想一次就列出下列三個變數:'H/R','W/E','W/L'的次數及相對次數分配表
a=['H/R','W/E','W/L']
for m in range(len(a)):
  f=data.groupby(a[m])
  percentage=f.size()/len(data)
  plist=pd.DataFrame(percentage,columns=["Percentage"])
  print(plist)
```

# 兩屬量變數分析：相關分析

方法一: 利用numpy之相關性corr相關分析的語法

```
#計算3P及PTS的相關係數
import numpy as np
np.corrcoef(data["3P"],data["PTS"])
```

```
#同時想一次進行很多變數的相關分析
#量的資料
A=['DIFF', '1Q',
       '2Q', '3Q', '4Q', 'F', 'MIN', 'FG', 'FGA', '3P', '3PA', 'FT', 'FTA',
       'OR', 'ORR', 'DRR', 'FT%', 'TOV%', 'EFG%', 'DRB%', 'ORB%', 'DR', 'TOT',
       'A', 'PF', 'ST', 'TO', 'BL', 'PTS', 'POSS', 'PACE', 'OEFF', 'DEFF',
       'REST DAYS', 'OPENING ODDS', 'SPREAD']
```

方法二: 利用scipy之相關性corr相關分析的語法

```
#計算PTS(總分)與其他變數的相關係數，並列出各變數之相關係數值
#皮爾森相關分析(量*量)
name=[]
alist=[]
for i in range(len(A)):
  (r,pv)=stats.pearsonr(data['PTS'],data[A[i]]) #(r, pv)，r為相關係數，pv為p-value
  name.append(A[i])
  alist.append(r)
```

```
#資料結構以DataFrame顯示 
#語法:pd.DataFrame(資料,index名稱,columns名稱)
result=pd.DataFrame(alist,name,columns=['與PTS之相關係數'])
result
```

# 兩類別變數分析：交叉分析

```
#語法:pd.crosstab(rows,columns,margins=True)，margins=True表顯示all，False或不命令margins則不顯示
#請計算"W/L" VS "REST DAYS" 的交叉次數表
pd.crosstab(data["W/L"],data["REST DAYS"],margins=False)
```

```
#normalize=True是指每個cell除以總數所得的比例;normalize=0是指每個cell除以列的總數所得的比例;normalize=1是指每個cell除以行的總數所得的比例;
pd.crosstab(data["W/L"],data["REST DAYS"],margins=False, normalize=1) 
```

```
#計算各球隊(TEAMS)在不同休息天數(REST DAYS)下勝場數(WL)之差異(利用groupby功能)
a=data.groupby(['TEAMS','REST DAYS']).sum()
a[["W/L"]]
```

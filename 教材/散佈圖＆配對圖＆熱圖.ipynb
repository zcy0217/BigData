```
#載入執行程式所需之套件
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#設定繪圖桌布大小
plt.rcParams['figure.figsize']=(5.0,5.0)
```

```
#讀取檔案，將檔案名稱設為data
data=pd.read_excel('nba2.xlsx')
```

```
data.head()
```

# 兩變數關係之圖形：散佈圖、配對圖與熱圖

```
sns.set(style='ticks',color_codes=True) #底色
```

散佈圖(sactter plot)之繪製:     
例子:請畫出'OEFF'和'PTS'的散佈圖(scatter plot)

方法一:採用matplot進行散佈圖之繪製(plt.scatter)

```
#alpha表透明度0~1、s表點大小
# colorlist=['red','pink','yellow','orange','skyblue']
plt.scatter(x=data["OEFF"],y=data["PTS"],color='b',s=10,alpha=1)
```

```
plt.scatter(x=data.OEFF,y=data.PTS,color='b',s=20,alpha=1)
```

```
#採用seaborn進行散步圖及回歸線(regplot)之繪製
#例子:請畫出'OEFF'和'PTS'的散步圖(scatter plot)及回歸線(regplot)之繪製
#sns.regplot可同時畫出散佈圖及回歸線
#回歸直線fit_reg、散佈圖scatter
sns.regplot(x=data.OEFF,y=data.PTS,color='b',scatter=True,fit_reg=False)
```

```
#採用seaborn進行散步圖及回歸線(regplot)之繪製
#例子:請畫出'OEFF'和'PTS'的散步圖(scatter plot)及回歸線(regplot)之繪製
#sns.regplot可同時畫出散佈圖及回歸線
#散佈圖scatter_kws、回歸直線line_kws
sns.regplot(x=data.OEFF,y=data.PTS,color='b',scatter_kws={"color":"b"}, line_kws={"color":"r"})
```

```
#用seaborn畫出pairplot(配對圖)
import seaborn as sns; sns.set(style='ticks')
pairlist=['PTS','OEFF','TO','DEFF']
sns.pairplot(data[pairlist])
```

```
#算相關係數、畫熱圖(heatmap)、顯示中間相關係數值（annot）
pairlist=['PTS','OEFF','TO','DEFF']
sns.heatmap(data[pairlist].corr(),annot=True,cmap='Blues')
```

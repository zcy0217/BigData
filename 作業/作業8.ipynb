請利用宜蘭生活機能.xlsx檔案資料分析治安"Total_secu", 交通"Total_traf",醫療"Total_phar",教育"Total_scho"等資源對房價“Total_Price”的影響。

```python
import pandas as pd
```

```python
data=pd.read_excel('宜蘭生活機能.xlsx')
data.head()
```

```python
X=data[["Total_secu","Total_traf","Total_phar","Total_scho"]] #自變數
Y=data["Total_Price"]
import statsmodels.api as smf
RA=smf.OLS(Y,X).fit()
print(RA.summary())
```

第五題

請分析哪些量化因素與房價具有較高的關聯性？

```
data2.columns
```

```
A=['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS',
       'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE']
```

```
name=[]
alist=[]
for i in range(len(A)):
  (r,pv)=stats.pearsonr(data2['PRICE'],data2[A[i]]) 
  name.append(A[i])
  alist.append(r)
```

```
result=pd.DataFrame(alist,name,columns=['與PRICE之相關係數'])
result
```

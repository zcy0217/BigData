```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='ticks',color_codes=True)

data=pd.read_excel('boston_house(type).xlsx')
data.head()
```

第一題

請畫出此資料中量化指標與房價(PRICE)相關性的配對圖

```python
import seaborn as sns
sns.set(style='ticks')
plt.rcParams['figure.figsize']=(5.0,5.0)
pairlist=['CRIM','ZN','INDUS','NOX','RM','AGE','PRICE']
sns.pairplot(data[pairlist])
```

```python
pairlist=['DIS','RAD','TAX','PTRATIO','B','LSTAT','PRICE']
sns.pairplot(data[pairlist])
```

第二題

請畫出此資料中量化指標與房價(PRICE)相關性的熱圖

```python
plt.rcParams['figure.figsize']=(10,10)
pairlist=['CRIM','ZN','INDUS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','PRICE']
sns.heatmap(data[pairlist].corr(),annot=True,cmap='Blues')
```

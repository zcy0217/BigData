作業二：請針對所提供的資料檔案(boston_house(type),xlsx)進行以下分析：

1.請畫出此資料中房價的直方圖。

2.請畫出此資料中房價超過20(PRICE TYPE=2)與低於20(PRICE TYPE=1)的個數的長條圖。

3.請畫出此資料中RAD各類別個數的圓餅圖。

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('boston_house(type).xlsx')
data.head()
```

第一題

請畫出此資料中房價的直方圖

```python
plt.rcParams['figure.figsize']=(5.0,5.0)
sns.distplot(data['PRICE'].dropna(),bins=10,kde=False)
```

第二題

請畫出此資料中房價超過20(PRICE TYPE=2)與低於20(PRICE TYPE=1)的個數的長條圖

```python
sns.set(style='ticks')
result=[sum(data['PRICE TYPE']==1),sum(data['PRICE TYPE']==2)]
sns.barplot(x=['PRICE TYPE=1','PRICE TYPE=2'],y=result)
```

第三題

請畫出此資料中RAD各類別個數的圓餅圖

```python
data.groupby('RAD').size()
```

```python
plt.rcParams['figure.figsize']=(9,9)
plt.rcParams['figure.facecolor']='white'
fig, axes=plt.subplots()
axes.pie([sum(data['RAD']==1),sum(data['RAD']==2),sum(data['RAD']==3),sum(data['RAD']==4),sum(data['RAD']==5),sum(data['RAD']==6),sum(data['RAD']==7),sum(data['RAD']==8),sum(data['RAD']==24)],labels=['1','2','3','4','5','6','7','8','24'],autopct='%.2f%%')
plt.legend(labels=['1','2','3','4','5','6','7','8','24'])
plt.show()
```

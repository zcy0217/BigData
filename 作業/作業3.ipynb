```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('鄉鎮市ATM_綜所稅.xlsx')
data.head()
```

第一題

畫出六都ATM數量、綜合所得稅總額(平均數)之比較圖

```python
a=data.groupby("COUNTYNAME")
b=a['綜合所得總額'].mean()
c=pd.DataFrame(b,columns=['綜合所得總額'])
c=round(c,2)
d=c.loc[['臺北市','新北市','桃園市','臺中市','臺南市','高雄市']]

a2=data.groupby("COUNTYNAME")
b2=a2['Total_ATM'].sum()
c2=pd.DataFrame(b2,columns=['Total_ATM'])
c2=round(c2,2)
d2=c2.loc[['臺北市','新北市','桃園市','臺中市','臺南市','高雄市']]

d.merge(d2,on = 'COUNTYNAME')
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(7,7)
x = d.index
y1 = d['綜合所得總額']
y2 = d2['Total_ATM']
fig = plt.figure() #將圖形名稱設為fig
plt.title('六都綜合所得稅總額、ATM數量之比較圖',fontsize=15)
plt.xlabel('六都',fontsize=10)
plt.ylabel('數量',fontsize=10)
plt.plot(x, y1, 'b-o',label="綜合所得稅總額")
plt.plot(x,y2,'r-o',label="ATM數量")
plt.legend(loc = "best",fontsize=15)
```

第二題

畫出臺北市ATM數量、綜合所得稅總額之盒形圖

```python
plt.rcParams['figure.figsize']=(7,7)
```

```python
a=pd.DataFrame(data)
b=a[a['COUNTYNAME'].isin(['臺北市'])] 
c=b['Total_ATM']
d=pd.DataFrame(c)
sns.boxplot(data=d,color='y')
plt.xlabel('臺北市ATM',fontsize=15)
plt.ylabel('ATM數量',fontsize=15)
```

```python
a2=pd.DataFrame(data)
b2=a2[a2['COUNTYNAME'].isin(['臺北市'])]
c2=b2['綜合所得總額']
d2=pd.DataFrame(c2)
sns.boxplot(data=d2,color='g')
plt.xlabel('臺北市綜合所得稅',fontsize=15)
plt.ylabel('金額',fontsize=15)
```

第四題

請畫出台灣一週本土與境外移入的疫情趨勢圖

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(5,5)
x = [18,19,20,21,22,23,24]
y1 = [1,6,3,5,10,2,15]
y2=[74,120,118,93,79,95,124]
plt.title('Trend of COVID 2022, 境內&境外 ',fontsize=20)
plt.xlabel('Date',fontsize=10)
plt.ylabel('Number of covid_19',fontsize=10)
plt.plot(x, y1, 'b-o',label="境內")
plt.plot(x,y2,'r-o',label="境外") 
plt.legend(loc = "best",fontsize=15)
```
